```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('bhavdata.csv') #Reading the CSV
df
```

```python
df.shape # Tells the Number of Rows and Columns present in the whole CSV File
```

```python
df1 = df[df['INSTRUMENT'].str.contains('OPTIDX')] #Filtering only OPT from the whole Instruments
df1
```

```python
# Define all the symbols that you want the OPTINX and OPTSTK for in the multisymbols, in the same format
#Filtering all the symbols to a new dataframe
multisymbols = ['NIFTY']
df1 = df1[df['SYMBOL'].isin(multisymbols)]
df1
```

```python
#Change according to liking
#Get a single date or a specific date
dates = ["27-Aug-2020"] 
df1 = df1[df['EXPIRY_DT'].isin(dates)]
df1
```

```python
#Filters out all the Values of only PE
df1 = df1[df['OPTION_TYP'].str.contains("PE")]
df1
```

```python
df1 = df1.sort_values(['CHG_IN_OI'], ascending=False)
df1
```

```python
df1 = df1.iloc[:1]
df1
```

```python
df1.to_csv('OPTIDX_PE_Nifty.csv')
```
